### 9. Создать одномерный массив а, состоящий из последовательности целых чисел от 0 до 12. Получить матрицу B, умножив матрицу A на матрицу At с помощью матричного умножения. Какой размер имеет матрица B? Получится ли вычислить обратную матрицу для матрицы B и почему?

```python
a = np.arange(0, 12)
print(a)
print()

A = a.reshape(4,3)
print(A)
print()


At = A.T
print(At)
print()

B = A.dot(At)
print(B)
print()
print(f'Размер матрицы В: {B.shape}')

d = np.linalg.det(B)
print(f'Определитель квадратной матрицы d: {d}')
if d == 0:
    print("Вычисление обратной матрицы невозможно при определителе равном 0")
```
